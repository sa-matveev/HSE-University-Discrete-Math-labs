# Работа с графами из файлов

В этом ноутбуке мы рассмотрим работу с большими графами

```python
# Загружаем граф из файла

import networkx as nx

G = nx.read_edgelist("amazon0302.txt", create_using=nx.Graph(), nodetype=int, data=False)
```

```python
# Считаем вершины и ребра

nodes = G.number_of_nodes()
print("Число вершин: ", nodes)

edges = G.number_of_edges()
print("Число ребер: ", edges)
```

```python
# Печатаем вершины

G.nodes()
```

```python
# Перенумеруем вершины

G = nx.convert_node_labels_to_integers(G)
```

```python
# Печатаем вершины

G.nodes()
```

```python
# Смотрим на окрестности вершин

list(G[50])
```

```python
# Рисуем фрагменты графа

import matplotlib.pyplot as plt

def draw_G(G):
    options = {
        "node_size": 3,
        "node_color": "black",
        "edgecolors": "black",
        "linewidths": 1,
        "width": 1,
    }

    nx.draw_networkx(G, with_labels=False, **options)
    plt.show()
    
draw_G(G.subgraph(G[300]))
```

```python
# Конвертируем в словарь списков

adjlist = nx.convert.to_dict_of_lists(G)

dict(list(adjlist.items())[0:10])
```

```python
# Конвертируем из словаря списков

G = nx.from_dict_of_lists(adjlist)

draw_G(G.subgraph(G[300]))
```
